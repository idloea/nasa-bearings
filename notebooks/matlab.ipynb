## Imports

```python
from pathlib import Path
from pymatreader import read_mat
import matplotlib.pyplot as plt

```

## Data

```python
DATA_PATH = Path('data/MFPT Fault Data Sets')

all_mat_files = sorted(DATA_PATH.rglob('*.mat'))
number_of_matlab_files = len(all_mat_files)
print(f'number_of_matlab_files: {number_of_matlab_files}')
```

```python
for mat_file in all_mat_files:
    data = read_mat(mat_file)
    plt.figure(figsize=(12, 3))
    plt.plot(data['bearing']['gs'])
    plt.title(f'Signal from file: {mat_file.name}')
    plt.grid()
    plt.show();
```
